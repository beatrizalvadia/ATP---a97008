## Leitura (exercício 1)

Depois de pensares num modelo adequado especifica a função que lê o ficheiro e povoa esse modelo.

Nota: os valores do campo id são demasiado complexos para serem usados na aplicação pelo utilizador.

Ideia: Cria ids novos, mais simples, tipo: emd00001, emd00002, ..., emd00100, ...

```python
# BdEMD = [EMD]
# EMD = [...]
# Leitura/carregamento da informação do ficheiro

def getEMD(linha):
    novaLinha=linha.replace("\n","")
    emd=[]
    campos=novaLinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range(2,len(campos)):
        emd.append(campos[i])
    return emd
    


def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline()
    for linha in f:
        bd.append(getEMD(linha))
    return bd

BD = lerDataset("emd.csv")
print(BD)
```

## Modalidades (exercício 4)

Especifica um função que dá como resultado uma lista de modalidades ordenada alfabeticamente e sem repetições.

```python
def modalidades(bd):
    l=[]
    for linha in bd:
        if linha[7] not in l:
            l.append(linha[7])
    l.sort()
    return l
    
print(modalidades(BD))
```

## Distribuição por modalidade (exercício 5)

Especifica uma função que dá como resultado uma lista de pares indicando quantos EMD estão registados em cada modalidade.

```python
def distribPorModalidade(bd):
    distribuicao={}
    for emd in bd:
        if emd[7] in distribuicao.keys():
            distribuicao[emd[7]]= distribuicao[emd[7]]+1
        else:
            distribuicao[emd[7]]= 1
    return distribuicao

print(distribPorModalidade(BD))
```

## Distribuição por ano (exercício 7)

Especifica uma função que dá como resultado uma lista de pares indicando quantos EMD estão registados por cada ano.

```python
def distribPorAno(bd):
    distribuicao={}
    for emd in bd:
        if emd[1][0:4] in distribuicao.keys():
            distribuicao[emd[1][0:4]]= distribuicao[emd[1][0:4]]+1
        else:
            distribuicao[emd[1][0:4]]= 1
    return distribuicao

print(distribPorAno(BD))
```

## Distribuição parametrizada (exercício 8)

Especifica uma função que permita calcular uma distribuição por qualquer um dos campos da BD.

```python
def distribPorIdade(bd):
    distribuicao={}
    for emd in bd:
        if emd[4] in distribuicao.keys():
            distribuicao[emd[4]]= distribuicao[emd[4]]+1
        else:
            distribuicao[emd[4]]= 1
    return distribuicao

def distribPorGenero(bd):
    distribuicao={}
    for emd in bd:
        if emd[5] in distribuicao.keys():
            distribuicao[emd[5]]= distribuicao[emd[5]]+1
        else:
            distribuicao[emd[5]]= 1
    return distribuicao

def distribPorFederado(bd):
    distribuicao={}
    for emd in bd:
        if emd[10] in distribuicao.keys():
            distribuicao[emd[10]]= distribuicao[emd[10]]+1
        else:
            distribuicao[emd[10]]= 1
    return distribuicao

def distribPorResultado(bd):
    distribuicao={}
    for emd in bd:
        if emd[11] in distribuicao.keys():
            distribuicao[emd[11]]= distribuicao[emd[11]]+1
        else:
            distribuicao[emd[11]]= 1
    return distribuicao

def distrib(bd,opcao):
    if opcao=="ano":
        return distribPorAno(bd)
    elif opcao=="clube":
        return distribPorClube(bd)
    elif opcao=="modalidade":
        return distribPorModalidade(bd)
    elif opcao=="idade":
        return distribPorIdade(bd)
    elif opcao=="genero":
        return distribPorGenero(bd)
    elif opcao=="federado":
        return distribPorFederado(bd)
    elif opcao=="resultado":
        return distribPorResultado(bd)

print(distrib(BD,"ano"))
print(distrib(BD,"clube"))
print(distrib(BD,"modalidade")) 
print(distrib(BD,"idade"))
print(distrib(BD,"genero"))
print(distrib(BD,"federado"))    
print(distrib(BD,"resultado"))
```

# Gráficos

## Gráfico da distribuição por modalidade (exercício 8)

Especifica uma função que faz o plot dum gráfico com a distribuição de alunos por modalidade.

```python
import matplotlib.pyplot as plt

def plotDistribPorModalidade(bd):
    plt.rcParams["figure.figsize"] = (20,8)
    left = []
    height = []
    tick_label = []
 
    a=distribPorModalidade(bd)
    i=1
    for elem in a:
        tick_label.append(elem)
        height.append(a[elem])
        left.append(i)
        i=i+1
    plt.bar(left, height, tick_label = tick_label,
        width = 0.8, color = ['red'])
    
    plt.xlabel('Modalidades')
    plt.ylabel('Número de exames')
    plt.title("Distribuição de alunos por modalidade")
    plt.show()
    
plotDistribPorModalidade(BD)
```
