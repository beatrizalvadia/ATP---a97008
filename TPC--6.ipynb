```python
import matplotlib.pyplot as plt

def criarTermo(coef,exp):
    return (coef, exp)

def criarPolinomio():
    return []

def inserirTermo(p, t):
    p.append(t)
  
def verTermo(t): 
    c, e = t
    if e == 0:
        return str(c)
    elif c == 1:
        return "x^" + str(e)
    else:
        return str(c) + "x^" + str(e)
    
def verPolinomio(p):
    politexto = ""
    if len(p) != 0:
        politexto = verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0] >= 0:
                politexto = politexto + "+" + verTermo(p[i])
            else:
                politexto = politexto + verTermo(p[i])
    return politexto

def grauPol(p):
    grau = p[0][1]
    for i in range(1,len(p)):
        if p[i][1] > grau:
            grau = p[i][1]
    return grau

def calcPol(p, x):
    valor = 0
    for (c, e) in p:
        valor = valor + c * x**e
    return valor

def tabela(p, linhas):
    print(" x " + " :: " + verPolinomio(p))
    for l in range(0, linhas):
        print(str(l) + " :: " + str(calcPol(p, l)))

def derivarPol(p):
    derivada = []
    for (c, e) in p:
        if e > 0:
            derivada.append((c*e, e-1))
    return derivada  

def chaveordenar(t):
    c,e=t
    return e

def ordenarpolinomio(p):
    p.sort(reverse=True, key=chaveordenar)
    return p

def simplificarPol(p):
    p = ordenarpolinomio(p)
    p1 = criarPolinomio()
    for t in p:
        t1=[t[0],t[1]]
        inserirTermo(p1,t1)
    i=0
    while i<len(p1)-1:
        if p1[i][1]==p1[i+1][1]:
            c=p1[i][0]+p1[i+1][0]
            t=criarTermo(c,p1[i][1])
            inserirTermo(p1,t)
            p1.pop(i)
            p1.pop(i)
            p1=ordenarpolinomio(p1)
            i=0
        else:
            i=i+1
    return p1

p=[(1,3), (-2,2), (1,0)]

def graficopolinomio(p):
    x = [n for n in range(-100,100)]
    y = []
    for n in x:
        y.append(calcPol(p,n))

    plt.plot(x, y)
    eixoXx = [-100,100]
    eixoXy = [0,0]

    plt.plot(eixoXx, eixoXy, color='red', linewidth = 1)

    eixoYx = [0,0]
    eixoYy = [-200,200]

    plt.plot(eixoYx, eixoYy, color='black', linewidth = 1)
    
    plt.ylim(-50,50)
    plt.xlim(-50,50)
    
    plt.xlabel('Abcissas')
    plt.ylabel('Ordenadas')
    
    plt.title('O Polinómio')
    plt.show()
    
def menu():
    opcao=10
    while opcao!=0: 
        print("""
        Menu:
        1) Criar polinómio
        2) Listar polinómio 
        3) Mostrar grau de polinómio
        4) Calcular polinómio num ponto
        5) Calcular derivada de polinómio
        6) Simplificar polinómio
        7) Calcular tabela de valores
        8) Gráfico do polinómio
        0) Sair""")
    
        opcao=int(input("Insira a opção quer usar: "))
        
        if opcao==1:
            ntermos=int(input("Introduza o número de termos do polinómio: "))
            p=criarPolinomio()
            for j in range(ntermos):
                c=int(input("Insira o coeficiente: "))
                e=int(input("Diga o expoente: "))
                t=criarTermo(c,e)
                inserirTermo(p,t)
                verPolinomio(p)
            print(verPolinomio(p))
           
        elif opcao==2:
            print(p)
            
        elif opcao==3:
            print(grauPol(p))
            
        elif opcao==4:
            x=int(input("Diga o valor de x: "))
            print(calcPol(p,x))
            
        elif opcao==5:
            p=simplificarPol(p)
            print(verPolinomio(derivarPol(p)))
    
        elif opcao==6:
            print(verPolinomio(simplificarPol(p)))
            
        elif opcao==7:
            nlinhas=int(input("Diga o nº de linhas da tabela: "))
            print(tabela(p,nlinhas))

        elif opcao==8:
            graficopolinomio(p)            
 
        elif opcao==0:
            print("OFF")
            
        else:
            print("Opcao inválida")
                
menu()
```
